# Ejercicio 4




Crear dos funciones: calcular_area_circulo(radio) que reciba el radio como parmetro y devuelva el rea del crculo. calcular_perimetro_circulo(radio) que reciba el radio como parmetro y devuelva el permetro del crculo. Solicitar el radio al usuario y llamar ambas funciones para mostrar los resultados.

```python
PI = 3.14

def calcular_area_circulo(radio):
    return PI * radio ** 2

def calcular_perimetro_circulo(radio):
    return 2 * PI * radio

radio = float(input("Ingrese el radio del crculo: "))
area = calcular_area_circulo(radio)
perimetro = calcular_perimetro_circulo(radio)

print(f"El rea del crculo es: {area:.2f}")
print(f"El permetro del crculo es: {perimetro:.2f}")
```

# Ejercicio 5




Crear una funcin llamada segundos_a_horas(segundos) que reciba
una cantidad de segundos como parmetro y devuelva la cantidad
de horas correspondientes. Solicitar al usuario los segundos y mostrar el resultado usando esta funcin.

```python
def segundos_a_horas(segundos):
    return segundos / 3600

segundos = float(input("Ingrese la cantidad de segundos: "))
horas = segundos_a_horas(segundos)
print(f"{segundos} segundos equivalen a {horas:.2f} horas.")
```

# Ejercicio 6




Crear una funcin llamada tabla_multiplicar(numero) que reciba un
nmero como parmetro y imprima la tabla de multiplicar de ese
nmero del 1 al 10. Pedir al usuario el nmero y llamar a la funcin.

```python
def tabla_multiplicar(numero):
    print(f"Tabla de multiplicar de {numero}:")
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")

numero = int(input("Ingrese un nmero para mostrar su tabla de multiplicar: "))
tabla_multiplicar(numero)
```

# Ejercicio 7




Crear una funcin llamada operaciones_basicas(a, b) que reciba
dos nmeros como parmetros y devuelva una tupla con el resultado de sumarlos, restarlos, multiplicarlos y dividirlos. Mostrar los resultados de forma clara.

```python
def operaciones_basicas(a, b):
    suma = a + b
    resta = a - b
    multiplicacion = a * b

    if b != 0:
        division = a / b
    else:
        division = None

    return (suma, resta, multiplicacion, division)

a = float(input("Ingrese el primer nmero: "))
b = float(input("Ingrese el segundo nmero: "))
suma, resta, multiplicacion, division = operaciones_basicas(a, b)

print("Suma:", suma)
print("Resta:", resta)
print("Multiplicacin:", multiplicacion)
if division is not None:
    print("Divisin:", division)
else:
    print("Divisin: No se puede dividir por cero.")
```

# Ejercicio 8




Crear una funcin llamada calcular_imc(peso, altura) que reciba el
peso en kilogramos y la altura en metros, y devuelva el ndice de
masa corporal (IMC). Solicitar al usuario los datos y llamar a la funcin para mostrar el resultado con dos decimales.

```python
def calcular_imc(peso, altura):
    return peso / (altura ** 2)

peso = float(input("Ingrese su peso en kilogramos: "))
altura = float(input("Ingrese su altura en metros: "))

imc = calcular_imc(peso, altura)
print(f"Tu ndice de masa corporal (IMC) es: {imc:.2f}")
```

# Ejercicio 9




Crear una funcin llamada celsius_a_fahrenheit(celsius) que reciba
una temperatura en grados Celsius y devuelva su equivalente en
Fahrenheit. Pedir al usuario la temperatura en Celsius y mostrar el
resultado usando la funcin.

```python
def celsius_a_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius = float(input("Ingrese la temperatura en grados Celsius: "))
fahrenheit = celsius_a_fahrenheit(celsius)
print(f"{celsius}C equivalen a {fahrenheit:.2f}F")
```

# Ejercicio 10




Crear una funcin llamada calcular_promedio(a, b, c) que reciba
tres nmeros como parmetros y devuelva el promedio de ellos.
Solicitar los nmeros al usuario y mostrar el resultado usando esta
funcin.

```python
def calcular_promedio(a, b, c):
    return (a + b + c) / 3

num1 = float(input("Ingrese el primer nmero: "))
num2 = float(input("Ingrese el segundo nmero: "))
num3 = float(input("Ingrese el tercer nmero: "))

promedio = calcular_promedio(num1, num2, num3)
print(f"El promedio de {num1}, {num2} y {num3} es: {promedio:.2f}")
```
